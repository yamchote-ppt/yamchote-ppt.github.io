# load data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'class':np.random.choice(['A','B','C'], size=100),'value':np.random.randn(100), 'new_class': np.random.choice(['a', 'b', 'c'], size=100)})
df
```

```python
df = pd.get_dummies(df, columns=['class']).astype({"class_A": int, "class_B": int, "class_C": int})
df
```

```python
from sklearn.preprocessing import LabelEncoder

#step 1: initiate object
le = LabelEncoder()

#step 2: fit
le.fit(df['new_class'])

#step 3: transform (and save to dataframe)
df['new_class'] = le.transform(df['new_class'])
df
```
